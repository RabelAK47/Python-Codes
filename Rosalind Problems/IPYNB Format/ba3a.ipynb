```python
K = int(input())
DNA = input()
Ara = []
for i in range(len(DNA)):
    sub = DNA[i:i+K]
    if(len(sub) == K):
        Ara.append(sub)
ans = sorted(Ara)
for i in range(len(ans)):
    print(ans[i])
```

```python
def prefix(my_str):
    return my_str[:len(my_str)-1]

def suffix(my_str):
    return my_str[1:]

F = open('D:/CCCCC.txt', 'r')
lines = []
ans = ""
for L in F.readlines():
    lines.append(L.strip())

Len = len(lines)
dc = {}
for i in range(Len):
    for j in range(Len):
        if (i != j):
            if(suffix(lines[i]) == prefix(lines[j])):
                k = lines[i]
                v = lines[j]
                dc[k] = v
sorted(dc.items())
for key, val in dc.items():
    print(key, "->", val)
F.close()
```
